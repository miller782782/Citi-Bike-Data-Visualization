### 03 The Elbow technique

```python
# Define the range of the potential clusters in the data

num_cl = range(1, 10) 

# Define k-means clusters in the range assigned above

kmeans = [KMeans(n_clusters=i) for i in num_cl]
```

```python
# Define the range of the potential clusters in the data

num_cl = range(1, 10) 

# Define k-means clusters in the range assigned above

kmeans = [KMeans(n_clusters=i) for i in num_cl]
```

```python
# Create a score that represents a rate of variation for the given cluster option.

score = [kmeans[i].fit(df).score(df) for i in range(len(kmeans))] 

score
```

```python
# Nan values are not allowed so for now let's drop the mean_soft_coral and mean_trout_per_tow columns

df = df[['mean_live_coral', 'mean_dead_coral', 'mean_cots_per_tow']]
```

```python
df_2 = df[['mean_dead_coral', 'mean_cots_per_tow']]
```

```python
# Retry the k-means scores

score = [kmeans[i].fit(df).score(df) for i in range(len(kmeans))] 

score
```

```python
# Retry the k-means scores for df_2

score = [kmeans[i].fit(df_2).score(df_2) for i in range(len(kmeans))] 

score
```

```python
# Plot the elbow curve using PyLab.

pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

### 04 Deciding on the number of clusters
The optimal number of clusters seems to be 3 or maybe 4. We will go with 3 initially since this is where the curve begins to level off

### 05 k-means Clustering 

```python
# Create the k-means object

kmeans = KMeans(n_clusters = 3)
```

```python
# Fit the k-means object to the data

kmeans.fit(df)
```

### 06 Create a 'clusters' column

```python
# Make df not just a copy of a slice from a DataFrame

df = df.copy()
```

```python
# Add clusters columns then check the column and count the values

df['clusters'] = kmeans.fit_predict(df)
```

```python
df.head()
```

```python
df['clusters'].value_counts()
```

### 07 Visualisations using clustered data 

```python
# Plot the clusters for the "mean_dead_coral" and "mean_cots_per_tow" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['mean_cots_per_tow'], y=df['mean_dead_coral'], hue=kmeans.labels_, s=100) 

ax.grid(False)
plt.xlabel('Mean CoTS Per Tow') # Label x-axis.
plt.ylabel('Mean Dead Coral') # Label y-axis.
plt.show()
```

```python
# Plot the clusters for the "mean_live_coral" and "mean_cots_per_tow" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['mean_cots_per_tow'], y=df['mean_live_coral'], hue=kmeans.labels_, s=100) 

ax.grid(False)
plt.xlabel('Mean CoTS Per Tow') # Label x-axis.
plt.ylabel('Mean Live Coral') # Label y-axis.
plt.show()
```

```python
# Plot the clusters for the "mean_live_coral" and "mean_cots_per_tow" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['mean_dead_coral'], y=df['mean_live_coral'], hue=kmeans.labels_, s=100) 

ax.grid(False)
plt.xlabel('Mean Dead Coral') # Label x-axis.
plt.ylabel('Mean Live Coral') # Label y-axis.
plt.show()
```

### 08 Discuss the clusters

In the mean_dead_coral vs mean_live_coral cluster plot the clusters  make sense and appear to be largely decided by the coverage of live coral with three clusters one above the other.  The situation in the mean_cots_per_tow vs mean_live_coral is very similar with the same pattern emerging.  

In the mean_cots_per_tow vs mean_live_coral however the clusters are mixed together and it is not clear (to me) why that is or what the clusters represent.

### 09 Calculate descriptive statistics 

```python
# Rename the 0,1,2 with their colours from the graph

df.loc[df['clusters'] == 2, 'cluster'] = 'dark purple'
df.loc[df['clusters'] == 1, 'cluster'] = 'purple'
df.loc[df['clusters'] == 0, 'cluster'] = 'pink'
```

```python
df.groupby('cluster').agg({'mean_live_coral':['mean', 'median'], 
                         'mean_dead_coral':['mean', 'median'], 
                         'mean_cots_per_tow':['mean', 'median']})
```

The purple clusters shows reefs which have the highest mean live coral (48%) with low CoTS density (0.14 per tow) and a slightly higher percentage of dead coral but still low overall.  These sites with their strong coral cover and low impact from CoTS means they are likely to be healthy, stable reefs.

The pink cluster shows reefs which have a medium coverage of live coral (27%) with low dead coral and low CoTS density.  These reefs are doing ok and while they are not as healthy as the purple reefs, they are not suffering major mortality or outbreaks either.

The dark purple cluster however shows reefs with the lowest live coral coverage at only 12% on average.  There is a moderately high CoTS presence at 0.46 per tow and a similar proportion of dead coral. These reefs are degraded and possibly recovering or could be experiencing ongoing coral loss.
