### Data cleaning and consistency checks

```python
df.describe()
```

```python
#  Add numeric midpoints to 'MEDIAN_SOFT_CORAL'


# Define the mapping from code to numeric midpoint (%)
midpoint_map = {
    '0': 0,
    '1L': 2.5,
    '1U': 7.5,
    '2L': 15,
    '2U': 25,
    '3L': 35,
    '3U': 45,
    '4L': 56.25,
    '4U': 68.75,
    '5L': 81.25,
    '5U': 93.75
}

# Add a new column to the data frame with mapped midpoints
df['MEDIAN_SOFT_CORAL_MIDPOINT'] = df['MEDIAN_SOFT_CORAL'].map(midpoint_map).astype(float)

# Replace any missing codes with NaN
df['MEDIAN_SOFT_CORAL_MIDPOINT'] = df['MEDIAN_SOFT_CORAL_MIDPOINT'].fillna(np.nan)

# Display the updated DataFrame
print(df.head())
```

```python
# Repeat for 'MEDIAN_LIVE_CORAL' and 'MEDIAN_DEAD_CORAL'

df['MEDIAN_LIVE_CORAL_MIDPOINT'] = df['MEDIAN_LIVE_CORAL'].map(midpoint_map).astype(float)
df['MEDIAN_LIVE_CORAL_MIDPOINT'] = df['MEDIAN_LIVE_CORAL_MIDPOINT'].fillna(np.nan)

df['MEDIAN_DEAD_CORAL_MIDPOINT'] = df['MEDIAN_DEAD_CORAL'].map(midpoint_map).astype(float)
df['MEDIAN_DEAD_CORAL_MIDPOINT'] = df['MEDIAN_DEAD_CORAL_MIDPOINT'].fillna(np.nan)
```

```python
df.head()
```

```python
# Convert all column names to lowercase

df.columns = df.columns.str.lower()
```

```python
df['p_code'].unique()
```

```python
# Probably we don't need the p_code column.  This may be deleted later.
```

```python
# Check data types for each column

print(df.dtypes)
```

```python
# Check for any columns of mixed type data

for col in df.columns.tolist():
  weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Change 'median_soft_coral' to type string

df['median_soft_coral'] = df['median_soft_coral'].astype('str')
```

```python
# Re-run the check for mixed type data

for col in df.columns.tolist():
  weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# There are now no mixed type columns 
```
