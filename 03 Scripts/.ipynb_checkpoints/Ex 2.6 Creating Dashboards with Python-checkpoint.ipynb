```python
df = df[['start_time', 'end_time','trip_duration',
        'start_station_name', 'end_station_name', 'start_lat','start_lng',
        'end_lat', 'end_lng', 'avgTemp', 'no_of_trips', 'date', 'day_of_week', 'day_type', 'start_hour']]
```

```python
# Create a month column

df['month'] = df['date'].dt.month
df['month'] = df['month'].astype('int')
```

```python
# Create a season column

df['season'] = ["winter" if (month == 12 or 1 <= month <= 4)
    else "spring" if (4 < month <= 5)
    else "summer" if (6 <= month <= 9)
    else "fall"
for month in df['month']
    ]
```

### 3. Create bar chart of top 20 more popular start stations 

```python
# Group data to find 20 most popular start stations

df['value'] = 1 
df_groupby_bar = df.groupby('start_station_name', as_index=False).agg({'value': 'sum'})
top20 = df_groupby_bar.nlargest(20, 'value')
```

```python
top20
```

```python
# Create bar chart of top 20 stations using plotly.graph_objects

fig = go.Figure(go.Bar(x = top20['start_station_name'], y = top20['value'], marker={'color': top20['value'],'colorscale': 'Reds'}))
fig.show()
```

```python
# Adding settings to customize the design on the graph

fig.update_layout(
    title = 'Top 20 most popular bike stations in New York',
    xaxis_title = 'Start stations',
    yaxis_title = 'Sum of trips',
    plot_bgcolor="#2b2b2b",   # inside axes
    paper_bgcolor="#2b2b2b",  # outside axes
    font=dict(color="white"),
    margin = dict(b=120),
    width = 900, height = 400)

fig.update_xaxes(
    showgrid = True,
    gridcolor="#444",
    zerolinecolor="#444")

fig.update_yaxes(
    showgrid = True,
    gridcolor="#444",
    zerolinecolor="#444")
```

```python
# This looks better (to me at least).  I prefer the dark background and it makes the lighter 
# bars more visible while retaining the emphasis on the darker bars.
```

### 4. Create dual-axis line chart for number of trips and average temperature 

```python
df.columns
```

```python
df.groupby('date')[['avgTemp', 'no_of_trips']].nunique().max()
```

```python
# Create data with only required columns to speed up creation of the line chart.

df_daily = df[['date', 'avgTemp', 'no_of_trips']].drop_duplicates()
```

```python
df_daily
```

```python
# Line chart

fig = make_subplots(specs = [[{"secondary_y": True}]])

fig.add_trace(
go.Scatter(x = df_daily['date'], 
           y = df_daily['no_of_trips'], 
           name = 'Daily bike rides',
           line=dict(color='palevioletred')),
    secondary_y = False, 
)

fig.add_trace(
go.Scatter(x=df_daily['date'], 
           y = df_daily['avgTemp'], 
           name = 'Daily temperature', 
           line=dict(color='orange')),
    secondary_y=True, 
)
```

```python
# Now customize the graph to fit the chosen theme

fig.update_layout(
    title = 'Average Temperature and Number of Rides per Day',
    plot_bgcolor="#2b2b2b",   # inside axes
    paper_bgcolor="#2b2b2b",  # outside axes
    font=dict(color="white"),
    margin = dict(b=120),
    width = 900, 
    height = 400,
    showlegend=False)

# x-axis
fig.update_xaxes(
    title_text = 'Date',
    showgrid = True,
    gridcolor="#444",
    zerolinecolor="#444")

# Left y-axis (no of trips)
fig.update_yaxes(
    title_text = 'No of trips',
    title_font=dict(color='palevioletred'),
    tickfont=dict(color='palevioletred'),
    showgrid = False,
    gridcolor="#444",
    zerolinecolor="#444",
    secondary_y = False)

# Right y-axis (Temperature)
fig.update_yaxes(
    title_text = 'Average Daily Temperature °C',
    title_font=dict(color='orange'),
    tickfont=dict(color='orange'),
    showgrid = True,
    gridcolor="#444",
    zerolinecolor="#444",
    secondary_y = True)
```

While the graphs appear to follow the same general trend, caution must be taken when reading a dual-axis graph not to assume causation between the variables.
