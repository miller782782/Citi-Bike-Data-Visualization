# Calculating Seasonal Variations 

```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# import data

df = pd.read_csv("daily_trips_temp.csv")
```

```python
df.head()
```

```python
# Create month column

df['month'] = pd.to_datetime(df['date']).dt.month
```

```python
# Define month groups

winter_months = {11,12,1,2,3,4}
summer_months = {6,7,8,9}
spring = {5}
autumn = {10}
```

```python
# Map each row to a group

df['month_group'] = df['month'].apply(
    lambda m: "Nov-Apr" if m in winter_months
    else "Jun-Sep" if m in summer_months
    else "May" if m in spring
    else "Oct" if m in autumn
    else "Other")
```

```python
monthly_averages = (df.groupby("month_group", as_index=False)['no_of_trips'].mean())
```

```python
monthly_averages
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(
    data =monthly_averages,
    x = 'month_group',
    y = 'no_of_trips'
)
plt.show()
```
