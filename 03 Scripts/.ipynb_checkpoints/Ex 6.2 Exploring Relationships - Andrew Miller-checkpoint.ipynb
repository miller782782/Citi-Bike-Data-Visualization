### 3. Correlation matrix heatmap

```python
# Create correlation table

df_num.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn
corr = sns.heatmap(df_num.corr(), annot = True, ax = ax)

corr.figure.savefig(os.path.join(path, '04 Analysis', 'Visualisations', 'correlation_heatmap'), bbox_inches='tight')
```

We can see that aside from the relationship between the median and mean live, dead and soft coral measures which you would expect to be strongly correlated, the next most significant correlation is between mean_cots_per_tow and mean_dead_coral with an r value of 0.23. While this indicates that there may be a weak relationship between these variables, we will look more closely at these variables to determine if there is indeed evidence that the a greater number of cots per tow results in a higher percentage of dead coral

It may be necessary to look at group reefs by shelf or by area to identify relationships which do not show in this initial correlation heatmap.

The scattergraph does not really provide solid evidence of causation between the mean number of COTS per tow and the mean percentage of dead coral.  The outlier value creates the impression that this could be the case, but there are a lot of instances of a very low number of COTS and the mean_dead_coral ranging from 0 to over 50%.  there is also several cases of the mean dead coral being around 15% and the mean COTS per tow varying up to almost 60.

### 7. Revisit the questions from the previous exercise 

    2. What is the relationship between crown-of-thorns starfish (CoTS) abundance and coral cover? 

    a. To what extent does CoTS abundance correlate with live coral cover?

  The analysis so far shows no correlation between CoTS abundance and live coral cover however we do see there is a weak correlation between CoTS abundance and amount of dead coral.  This could be an indication that while CoTS do destroy coral and so many of them can destroy a reef, a lack of CoTS does not necessarily mean that the reef will have a high percentage cover of live (hard) coral.  There are clearly other factors which kill coral.  Hard coral reefs take decades to recover so once a reef is dead, it takes a long time before it would be observed to harbour a significant covering of live (hard) coral.

    3. How does coral trout abundance relate to coral health and/or CoTS density?

     a. Are trout more abundant on reefs with high coral cover?

There does not appear to be significant evidence that trout are more abundant on reefs with high live coral cover. Only a very weak correlation (r=0.12) exists between the variables.
