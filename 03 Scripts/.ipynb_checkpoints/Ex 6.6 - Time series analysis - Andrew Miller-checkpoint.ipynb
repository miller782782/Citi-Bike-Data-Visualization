### 06. Test for Stationarity 

We will start with the following hypotheses:

H0: There is a root unit present in the data which makes the data non-stationary

H1: There is no root unit present and so the data is stationay

```python
# Dickey-Fuller test

from statsmodels.tsa.stattools import adfuller 

def dickey_fuller(timeseries): # Define the function
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

# Apply the test using the function on the time series
dickey_fuller(df_lir['mean_live_coral'])

```

The test statistic is not smaller than any of the critical values so we cannot reject the null hypothesis.  The data is therefore non-stationary.

```python
# Plot of autocorrelations

# Import the autocorrelation and partial correlation plots
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf 
plot_acf(df_lir)
plt.show()
```

### 07. Differencing to make the data stationary

```python
# Take differences, t - (t -1)

data_diff = df_lir - df_lir.shift(1) 
```

```python
# Remove missing values that arose as a result of differencing

data_diff.dropna(inplace = True) 
```

```python
# Check out what the differencing did to the time-series curve

plt.close('all')
plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff)
plt.show()
```

```python
data_diff.columns
```

```python
# Rerun the Dickey-Fuller test

from statsmodels.tsa.stattools import adfuller 

def dickey_fuller(timeseries): # Define the function
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

# Apply the test using the function on the time series
dickey_fuller(data_diff['mean_live_coral'])
```

```python
# The test statistic is now lower than the critical value at the 5% level so we can now reject the null hypothesis and 
# consider the data to now be stationary.
```

```python
train = data_diff.loc[:'2020-01-01', 'mean_live_coral']
test = data_diff.loc['2020-01-01':, 'mean_live_coral']
```

#### Running and fitting the model 

```python
# Import the model we need

from statsmodels.tsa.api import ARIMA
```

```python
model = ARIMA(train, order=(5, 1, 1))
fitted = model.fit()
print(fitted.summary())
```

```python
# Get forecast object
forecast_obj = fitted.get_forecast(steps=len(test))

# Extract mean forecast and confidence intervals
fc_series = forecast_obj.predicted_mean
conf = forecast_obj.conf_int(alpha=0.05)

# Align indexes
fc_series.index = test.index
conf.index = test.index

# Use .iloc instead of [:, 0]
lower_series = conf.iloc[:, 0]
upper_series = conf.iloc[:, 1]

# Plot
plt.figure(figsize=(12,4), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.05)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

Both the graph and the P>|x| column confirm that the prediction was not very good.
