### 5. Recreate the dual-axis line plot from Ex 2.3 using seaborn

### 6. Boxplot of Members vs Casual users

```python
# Filter data for trips <= 200 mins and create boxplot of trip duration by member type.

df_plot = df[df['trip_duration'] <= 200]

sns.boxplot(x = 'member_casual', y = 'trip_duration', data = df_plot)
plt.show()
```

```python
# There shouldn't be negative values for trip_duration. Let's find out what is going on...
```

```python
df_plot.groupby('member_casual')['trip_duration'].describe()
```

```python
df[df['trip_duration']<0]
```

```python
# Count negative values of trip_duration

(df['trip_duration'] < 0).sum()
```

```python
# All of these negative values are caused by the clocks going back by 1 hour at 2am on November 6th 2022
# https://www.timeanddate.com/time/change/usa/new-york?year=2022

# Add 60 to each of these values to correct the duration
df.loc[df['trip_duration'] < 0, 'trip_duration'] += 60
```

```python
# Re-count negative values of trip_duration

(df['trip_duration'] < 0).sum()
```

```python
# Filter data for trips <= 200 mins and create boxplot of trip duration by member type.

df_plot = df[df['trip_duration'] <= 200]

boxplot_colors = {'member': '#52305D', 'casual': '#DDA88B'}
sns.boxplot(x = 'member_casual', y = 'trip_duration', data = df_plot, palette = boxplot_colors)
plt.show()
```

This is a much better box plot now that the nonsense negative durations have been fixed.

```python
# Recalculate descriptive statistics

df_plot.groupby('member_casual')['trip_duration'].describe()
```

The box plot allows us to compare the summary statistics for trip duration for members and for casual users of CitiBike.  We see that casual users had a higher media trip duration so they borrowed bikes for a longer time of almost 15 mins vs 9 mins for members.  Perhaps members are more frequent users and know exactly what purpose they need the bike for (commuting to work for instance) and so do not hang about whereas casual users perhaps are less regular in their journeys and take their time while using the bikes. The spread of the casual users is larger too with a greater interquartile range of almost 17 minutes compared to approx 11 minutes for members.  This shows that rides durations are less consistent for casual users.  Both box plots have many outliers at above the threshold so there are users of both types who use the bikes for longer duration journeys.

### 7. Create a Facetgrid 

I will create a Facegrid to show how weekend use differs from weekday usage. The side-by-side plots will be a good way to see the difference in the usage distributions whereas two lines on top of each other may get in each other's way and be less easy to read.

```python
df.head()
```

```python
# Check that 'date' and 'start_time' are both of type datetime

df['date'].dtype
```

```python
df['start_time'].dtype
```

```python
# Create a day-of-week column from 'date'

df['day_of_week'] = df['date'].dt.day_name()
```

```python
df.head()
```

```python
# Create a weekday/weekend flag

df.loc[df['day_of_week'].isin(['Saturday','Sunday']),'day_type'] = 'Weekend'

df.loc[df['day_of_week'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday']),'day_type'] = 'Weekday' 
```

```python
df.head()
```

```python
# Extract the hour of day from 'start_time'

df['start_hour'] = df['start_time'].dt.hour
```

```python
df.head()
```

```python
# Aggregate to daily hourly counts

df_hourly = (df.groupby(['date', 'day_type', 'start_hour']).size().reset_index(name='trip_count'))
df_hourly
```

```python
# Calculate the average day profile

df_avg_day = (df_hourly.groupby(['day_type', 'start_hour'])['trip_count'].mean().reset_index())
df_avg_day
```

```python
# Create FacetGrid showing average usage on weekdays vs weekend (line graph)

colors = {'Weekday': '#52305D', 'Weekend': '#DDA88B'}

g = sns.FacetGrid(df_avg_day, col='day_type', hue='day_type', palette = colors, height=3, aspect=1.2, sharex=True, sharey=True)

g.map_dataframe(sns.lineplot, x='start_hour',y='trip_count')

g.set_axis_labels("Hour of Day", "Average Number of Trips")
g.set_titles("{col_name}")
g.set(xlim=(0,23))

for ax in g.axes.flat:
    ax.set_xticks(range(0, 24, 2))

plt.tight_layout()
plt.show()
```

```python
# The lines look kind of unsatifying.  Maybe a bar chart (that resembles as histogram) would look better.

g = sns.FacetGrid(df_avg_day, col='day_type', hue='day_type', palette = colors, height=3, aspect=1.2, sharex=True, sharey=True)

g.map_dataframe(sns.barplot, x='start_hour', y='trip_count',edgecolor='#000000')

g.set_axis_labels("Hour of Day", "Average Number of Trips")
g.set_titles("{col_name}")
g.set(xlim=(-0.5,23.5))

for ax in g.axes.flat:
    ax.set_xticks(range(0, 24, 2))

plt.tight_layout()
plt.show()
```

The bar chart looks much more satisfying than the line graph somehow and I think better emphasises the difference in the distribution. 

This was in answer to question 4: "How does weekend usage compare to that during the week?"
We see that on weekdays there are clear peak demand spikes at 8-9am reaching an average of almost 6000 trips per hour and then again at 5-7pm surpassing 8000 trips per hour. Outside of the peak times demand drops. This strongly suggests that a lot of people are using the bikes to get to and from work.  What is interesting is that the demand in the evening is even higher than in the morning which implies that many people only use the bikes to get home from work but not in the morning.  

At the weekends the distribution does not have rush hour peaks but instead demand grows throughout the morning and early afternoon with the peak usage reaching 6000 trips per hour between 2 and 4pm but with usage remaining high throughout the day.
