### 4. Explore the data visually

### 5. Hypothesis

Sites with higher abundances of crown-of-thorns starfish (CoTS) will have higher mean dead coral cover.

H0 : There is no relationship between CoTS abundance and mean dead coral cover.

H1 : Mean dead coral cover increases with increasing CoTS abundance.

### 6. Reshape the variables 

```python
# Reshape the variables

X = df['mean_cots_per_tow'].values.reshape(-1,1)
y = df['mean_dead_coral'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

### 7. Split the data into a training set and a test set 

```python
# Split data into training set and test set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

### 8. Regression Analysis 

```python
# Create a regression object

regression = LinearRegression()
```

```python
# Fit the regression object to the training data for each variable

regression.fit(X_train, y_train)
```

```python
# Predict the values of y using X

y_predicted = regression.predict(X_test)
```

### 9. Plot showing regession line on the test set 

```python
# Create a plot that shows the regression line from the model on the test set.

plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('Mean COTS per tow vs Mean dead coral (Test set)')
plot_test.xlabel('Mean COTS per tow')
plot_test.ylabel('Mean Dead Coral')
plot_test.show()
```

```python
# Create objects that contain the model summary statistics, RMSE and R^2

rmse = mean_squared_error(y_test, y_predicted) 
r2 = r2_score(y_test, y_predicted)
```

### 10. Interpretation

The negative R2 score indicates just how poorly the model fits the data.
A negative score means that the model fits worse than a horizontal line.  

### 11. Check the model performance statistics 

```python
# Print the model summary statistics to evaluate the performance of the model.

print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```

### 13. Thoughts on how the model performed 

The model didn't perform well at all because the data really doesn't have much of a correlation in the first place.  I don't think it is even an issue of bias, I think that there are just other factors that contribute to dead coral besides just crown-of-thorns starfish. While a large outbreak of CoTS certainly concurrs with an increase in the amount of dead coral as we see in the outlier value, there are plenty of instances of dead coral without CoTS outbreaks which suggests that CoTS are not entirely responsible.
